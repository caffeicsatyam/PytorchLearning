# Reduction operation

```python
e = torch.randint(size=(2,3), low=0, high=10)
e
```

```python
# Sum
torch.sum(e) , torch.sum(e, dim=0), torch.sum(e,dim = 1 )
```

```python
# Mean 
e = e.type(torch.float16)
torch.mean(e), torch.median(e), torch.var_mean(e)
```

```python
# arg min and max
torch.argmax(e), torch.argmin(e)
```
